```python
import ee
import branca.colormap as cm
import folium
import foliumgee
from folium import plugins

ee.Initialize()

# 51.5074° N, 0.1278° W
lat = 51.5074
lon = -0.1278

image = ee.Image('JRC/GSW1_0/GlobalSurfaceWater')
m     = foliumgee.map(image)
base  = folium.Map()
mbox = folium.Map(tiles='https://api.mapbox.com/styles/v1/jsosa/cjj1ic27h0q3h2spkaegaufew/tiles/256/{z}/{x}/{y}?access_token=pk.eyJ1IjoianNvc2EiLCJhIjoiY2pqMWlhc2t4MG5xbjNwbndmY2R2MG90YiJ9.89W3SqG47t_6hyTNWWUjMg',attr='Mapbox')
vmin = 0
vmax = 100
step = 10
colormap = cm.linear.RdPu_09.scale(vmin, vmax).to_step(step)
colormap.caption = 'Water Occurrence (1984-2015)'
pale = ','.join([colormap.rgb_hex_str(j).replace('#','') for j in range(vmax)])
foliumgee.layer(mbox,image,vis_params={'bands':'occurrence','min':str(vmin),'max':str(vmax),'palette':pale,'opacity':'0.5'})
mbox.add_child(colormap)
mbox
```
